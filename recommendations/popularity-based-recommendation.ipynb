## Popularity-Based Filtering Recommendation System:

**Here we will display the top N highest rated books ( N = number of books)**

```python
Ratings
```

```python
Books
```

```python
name_ratings = pd.merge(Books, Ratings, on='ISBN')
number_ratings_df = name_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
number_ratings_df.rename(columns={'Book-Rating': 'Number-of-Ratings'}, inplace=True)
```

```python
number_ratings_df
```

```python
average_rating_df = name_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
average_rating_df.rename(columns={'Book-Rating': 'Average-Rating'}, inplace=True)
```

```python
average_rating_df
```

```python
popularity_df = pd.merge(number_ratings_df, average_rating_df, on='Book-Title')
```

```python
popularity_df
```

```python
popularity_df.sort_values('Average-Rating', ascending=False)
```

**Take the books that have rated 250 or more**

```python
popularity_df = popularity_df[popularity_df['Number-of-Ratings']>=250]
```

```python
popularity_df
```

```python
popularity_df = popularity_df.sort_values('Average-Rating', ascending=False)
```

```python
popularity_df
```

```python
popularity_df
```

**Display the full info for the most popular books**

```python
popularity_df = popularity_df.merge(Books, on='Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df
```

**The variables that we are going to use in the recommendation:**

```python
#ISBN,
# Book-Title, 
# Book-Author,
# Number-of-Ratings,
# Average-Rating,
# Image-URL-S, 
# Image-URL-M, 
# Image-URL-L
```

```python
popularity_df = popularity_df[['ISBN','Book-Title', 'Book-Author', 'Number-of-Ratings', 'Average-Rating', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L']]
```

```python
popularity_df
```

### We have above the most popular 186 books based on popularity

```python
def recommend_most_popular_books(num):
    return popularity_df.head(num)
```
